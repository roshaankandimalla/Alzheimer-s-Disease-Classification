# Univariate Analysis (categorical  Variables)
We visualize the distribution of each categorical column using barplot.

```python
categorical_cols=df.select_dtypes(include="int64").drop(['SystolicBP','DiastolicBP','Age'],axis=1).columns
```

```python
i=1
plt.figure(figsize=(20,20))
for col in categorical_cols[0:10] :
    plt.subplot(5,2,i)
    df[col].value_counts().plot(kind="bar",color = "lightblue", edgecolor = "black")
    i+=1
plt.tight_layout()
plt.show()
```

```python
i=1
plt.figure(figsize=(20,20))
for col in categorical_cols[10:] :
    plt.subplot(5,2,i)
    df[col].value_counts().plot(kind="bar",color = "lightblue", edgecolor = "black")
    i+=1
plt.tight_layout()
plt.show()
```

##  Bivariate Analysis (Numerical Feature vs Diagnosis)

Boxplots help compare distributions of numerical features for Alzheimer’s (1) vs Non-Alzheimer’s (0).

```python
i=1
plt.figure(figsize=(20,20))
for col in numerical_cols:
    if col != 'Diagnosis':
        plt.subplot(8,2,i)
        sns.boxplot(x='Diagnosis', y=col, data=df)
        plt.title(f"{col} vs Diagnosis")
        i+=1
        plt.grid()
plt.tight_layout()
plt.show()
```

#EDA–1: Standard Scaler + One-Hot Encoding

Import Required Libraries

```python
from sklearn.preprocessing import StandardScaler,OneHotEncoder
```

Define Numerical & Categorical Columns

Apply StandardScaler

```python
st=StandardScaler()
starray=st.fit_transform(df[numerical_cols])
st_df=pd.DataFrame(starray,columns=st.get_feature_names_out())
```

```python
st_df
```

#ONE-HOT ENCODING

Apply OneHotEncoder

```python
one=OneHotEncoder(sparse_output=False)
one_array=one.fit_transform(df[["Ethnicity","EducationLevel"]])
one_df=pd.DataFrame(one_array,columns=one.get_feature_names_out())
one_df
```

Combine Everything into Final Dataset

```python
eda1_df=pd.concat([st_df,one_df,df[categorical_cols].drop(["Ethnicity","EducationLevel"],axis=1)],axis=1)
```

View Final EDA-1 Dataset

```python
eda1_df.head()
```

```python
eda1_df.to_csv("eda1_df.csv",index=False)
```
